```python
import numpy as np
class perceptron:
    def __init__(self,learnspeed= 0.01,dataset=10):
        self.learnspeed=learnspeed
        self.dataset = dataset
    def fit(self,X,y) :
        self.weight = np.zeros(1 + X.shape[1])
        self.errors_= []
        for _ in range(self.dataset):
            errors = 0
            for xi,target in zip(X,y):
                update = self.learnspeed*(target-self.predict(xi))
                self.weight[1:] += update*xi
                self.weight[0] += update
                errors += int(update != 0.00)
                self.errors_.append(errors)
        return self
    def net_input(self,X):
        return np.dot(X,self.weight[1:] )+ self.weight[0]
    def predict(self,X):
        return np.where(self.net_input(X) >= 0.0,1,-1)

import pandas as pd
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/639388c2cbc2120a14dcf466e85730eb8be498bb/iris.csv')
b = df.tail()
b
import matplotlib.pyplot as plt
import numpy as np
y = df.iloc[0:100, 4].values
y = np.where(y== 'Iris-setosa',-1,1)
X = df.iloc[0:100,[0,2]].values
plt.scatter(X[:50,0],X[:50,1],color='red',marker='o',label='setosa')
plt.scatter(X[50:100,0],X[50:100,1],color='blue',marker='x',label='versicolor')
plt.xlabel('petallength')
plt.ylabel('sepal length')
plt.legend(loc='upper left')
plt.show()
machine = perceptron(learnspeed=0.1,dataset=10)
machine.fit(X,y)
plt.plot((range(1,len(machine.errors_)+1),machine.errors_),marker='o')
plt.xlabel('Epochs')
plt.ylabel('no of misclassifications')
plt.show()

```
